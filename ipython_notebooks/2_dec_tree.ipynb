<h1>Важность признаков - Decision Tree Classifier</h1>

```python
import pandas as pd
import numpy as np
from sklearn import tree
```

```python
path_to_data = 'D:\\WORK\ml\\data\\'
```

```python
data = pd.read_csv(path_to_data+'titanic_dataset2.csv', index_col='PassengerId')
```

```python
features = data.loc[:,['Survived', 'Pclass', 'Fare', 'Age', 'Sex']]
```

```python
features_flt = features.dropna(how='any', axis=0)
# features_flt = features[flt == False]
tgt_var = features_flt['Survived']
features_flt = features_flt.drop('Survived', 1)
features_flt['Sex'] = features_flt['Sex'].map(lambda s: s == 'female'); features_flt
```

```python
clf = tree.DecisionTreeClassifier(random_state=241)
for f, i in zip(features_flt.columns, 
                clf.fit(features_flt, tgt_var).feature_importances_):
    print(f"{f}: {i}")
```
